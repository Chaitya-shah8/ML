```python
import matplotlib.pyplot as plt
%matplotlib inline

x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'g*-')
```

```python
plt.plot(x,y,color='black',marker='+',linestyle='-',markersize=10)

#alpha is transparency
```

```python
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
day=[1,2,3,4,5,6,7]
```

```python
plt.xlabel("days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(day, max_t, label="max")
plt.plot(day, min_t, label="min")
plt.plot(day, avg_t, label="average")

plt.legend(loc="best")
plt.grid()
```

```python
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,label='Revenue')

plt.xticks(xpos,company)
plt.ylabel("Revenue")
plt.title("US tech stocks")
plt.legend()
```

```python
profit=[40,2,34,12]

plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2, profit,width=0.4,label="profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue")
plt.title("US tech stocks")
plt.legend(loc="best")
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.6, color=['blue','#FF007F'],label=['men','women'])
plt.legend()
```
